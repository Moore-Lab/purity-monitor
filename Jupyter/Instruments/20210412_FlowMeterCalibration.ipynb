```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use('../style.mplstyle')
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
Input = np.array([0,100,200,250,260,270,280,290,300,400,500,1000,1500,2000,2500,3000,3500,4000])
Output = np.array([-0.250,-0.261,-0.262,-0.266,-0.266,-0.267,-0.267,-0.267,0.07,0.1,0.123,0.248,0.373,0.498,0.623,0.748,0.873,0.997])*4000/1.32
```

```python
def linear(x,a,b):
    return a*x+b

start = 8
popt, pcov = curve_fit(linear, Input[start:], Output[start:])
x = np.linspace(0,4000,1000)
```

```python
plt.figure()
plt.xlabel('Input flow [sccm]')
plt.ylabel('Output flow [sccm]')
plt.xlim(0,4000)
plt.ylim(0,3500)

plt.plot(x, linear(x,*popt), color='r',  label='${:.6f}\cdot x + {:.6f}$'.format(popt[0], popt[1]))
Output[Output<0] = 0
plt.plot(Input, Output, color='k', marker='o', markersize=5, ls=':', label='Data')

plt.yticks(np.arange(0,3501,500))
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
