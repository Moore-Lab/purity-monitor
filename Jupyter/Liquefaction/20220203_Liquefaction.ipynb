```python
def numeric_temp(p_input):
    x=np.linspace(161,180,1000)
    p=10**(3.80675-(577.661/(x-13.0)))*1000 #pressure in mbar when x is in Kelvin
    t=np.interp(p_input,p,x)
    return t
```
